### **Task 01: Source the Data Set**

```python
from sklearn.datasets import load_iris

#loading iris dataset
iris_data = load_iris()

#variable names
keys = iris_data.keys()
features = iris_data['data']
features_shape = iris_data['data'].shape
target = iris_data['target']
target_shape = iris_data['target'].shape
target_names = iris_data['target_names']
features_names = iris_data['feature_names']

#The Iris dataset contains information on different samples of iris flowers, first used by Fisher
print(iris_data)
```

**Functions used throughout the notebook**  
Any created functions will be put here and called upon later throughout the notebook

```python
import numpy as np

#Functions for printing information on the iris dataset just for easier reading
def iris_info(name, var):
    print(f"\n{name} of the data:")
    print(var)

#functions for getting the first and last rows of a dataset for task02
#using variable n in case there's a need to get first or last 'n' number of rows instead of 5
def iris_firstrow_data(array, n=5):
    print(f"\nFirst {n} rows of data:")
    print(array[:n])

def iris_lastrow_data(array, n=5):
    print(f"\nLast {n} rows of data:")
    print(array[-n:])

#functions for summarizing the iris dataset for task 03
def iris_features_summary(features, features_names, stat=""):
    #making a dictiononary to pair each numpy function
    stat_function = {

        "mean" : np.mean,
        "min" : np.min,
        "max" : np.max,
        "std" : np.std,
        "median" : np.median

    }

    function = stat_function[stat]
    print(f"\n{stat.capitalize()} of each feature:")
    
    for r, name in enumerate(features_names):
        answer = function(features[:, r])
        print(f"{name}: {answer:.2f}")
```

### **Task 02: Explore the Data Structure**

```python
#The data has 150 iris flowers in it, with each having 4 features, as shown by the shape
iris_info("Features Shape", features_shape)

#This showcases the first 5 rows of data, and the last 5 rows of data
iris_firstrow_data(features)
iris_lastrow_data(features)

#The features are the sepal length and width, and the petal length and width (4 total)
iris_info("Features Names", features_names)

#The flowers are categorized into 3 different spieces: setosa, versicolor, virginica
iris_info("Target Names", target_names)
```

### **Task 03: Summarize the Data**

```python
#Summarizing the iris dataset (using numpy)

iris_features_summary(features, features_names, "mean")
iris_features_summary(features, features_names, "min")
iris_features_summary(features, features_names, "max")
iris_features_summary(features, features_names, "std")
iris_features_summary(features, features_names, "median")
```
