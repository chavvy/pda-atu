### **Task 01: Source the Data Set**

```python
from sklearn.datasets import load_iris
import pandas as pd

#loading iris dataset
iris_data = load_iris()

#variable names
keys = iris_data.keys()
features = iris_data['data']
features_shape = iris_data['data'].shape
target = iris_data['target']
target_shape = iris_data['target'].shape
target_names = iris_data['target_names']
features_names = iris_data['feature_names']

#The Iris dataset contains information on different samples of iris flowers, first used by Fisher
print(iris_data)
```

**Functions used throughout the notebook**  
Any created functions will be put here and called upon later throughout the notebook

```python
import numpy as np

#Functions for printing information on the iris dataset just for easier reading
def iris_info(name, var):
    print(f"\n{name} of the data:")
    print(var)

#functions for getting the first and last rows of a dataset for task02
#using variable n in case there's a need to get first or last 'n' number of rows instead of 5
def iris_firstrow_data(array, n=5):
    print(f"\nFirst {n} rows of data:")
    print(array[:n])

def iris_lastrow_data(array, n=5):
    print(f"\nLast {n} rows of data:")
    print(array[-n:])

#functions for summarizing the iris dataset for task 03
def iris_features_summary(features, features_names, stat=""):
    #making a dictiononary to pair each numpy function
    #wanted this to have similar functionality to the iris_info function
    #where one function can be used for a similar purpose by calling different arguments
    stat_function = {

        "mean" : np.mean,
        "min" : np.min,
        "max" : np.max,
        "std" : np.std,
        "median" : np.median

    }

    function = stat_function[stat]
    print(f"\n{stat.capitalize()} of each feature:")
    
    for r, name in enumerate(features_names):
        answer = function(features[:, r])
        print(f"{name}: {answer:.2f}")
```

### **Task 02: Explore the Data Structure**

```python
#The data has 150 iris flowers in it, with each having 4 features, as shown by the shape
iris_info("Features Shape", features_shape)

#This showcases the first 5 rows of data, and the last 5 rows of data
iris_firstrow_data(features)
iris_lastrow_data(features)

#The features are the sepal length and width, and the petal length and width (4 total)
iris_info("Features Names", features_names)

#The flowers are categorized into 3 different spieces: setosa, versicolor, virginica
iris_info("Target Names", target_names)
```

### **Task 03: Summarize the Data**

```python
#Summarizing the iris dataset (using numpy)

iris_features_summary(features, features_names, "mean")
iris_features_summary(features, features_names, "min")
iris_features_summary(features, features_names, "max")
iris_features_summary(features, features_names, "std")
iris_features_summary(features, features_names, "median")
```

### **Task 04: Visualize Features**

```python
#plotting a histogram for each feature in the data
import matplotlib.pyplot as plt

#making a loop to go through each column
for r in range(features.shape[1]):
    #plotting the histogram
    plt.hist(features[:, r], bins=15, alpha=0.6, color= "royalblue", edgecolor = "black", label = "Iris Flowers")

    #labels, titles, and legend
    plt.title(f"Histogram of {features_names[r]}")
    plt.xlabel(features_names[r])
    plt.ylabel("Frequency")
    plt.legend()

    #displaying the plot
    plt.show()

```

**Trying another method for task 04**  
Read about being able to "transpose" the data (switching the rows and columns)  
as well as "zip" (pairing the feature names with its data collumn) 

```python
for name, column in zip(features_names, features.T):
    #plotting the histogram
    plt.hist(column, bins=15, color= "lightgreen", edgecolor= "black", label = "Iris Flowers")

    #labels, titles, and legend
    plt.title(f"Histogram of {name}")
    plt.xlabel(name)
    plt.ylabel("Frequency")
    plt.legend()

    #displaying the histogram
    plt.show()
```

### **Task 05 and 06: Investigate Relationships, Analyze Relationships**    
Grouping task 05 and 06 together

```python
#getting the sepal length and petal length for plotting
x_sepal_length = features[:, 0]
y_petal_length = features[:, 2]

#setting the colors for each class
colors = ['red', 'green', 'blue']

#plotting the points for each class using a for loop and range()
#my understanding is that this for loop is used as a type of filter
#the classes are assigned a number:
#setosa = 0, versicolor = 1, virginica = 2
#the "target == target_class" matches the data of each flower to its respective target classification
#then it matches the color to the appropriate target, 0 = red, 1 = green, 2 = blue

for target_class in range(3):
    plt.scatter(
        x_sepal_length[target == target_class],
        y_petal_length[target == target_class],
        color = colors[target_class],
        label = target_names[target_class]
    )

#adding a line of regression using .polyfit()
#.polyfit() gives the slope and intercept of the regression line
#equation of a line: y = mx + C, m = slope, C = intercept
slope, intercept = np.polyfit(x_sepal_length, y_petal_length, 1)
line = (slope * x_sepal_length) + intercept
plt.plot(x_sepal_length, line, color = "skyblue", label = "Regression Line")

#labels, titles, and legend
plt.title("Sepal Length and Petal Length by Iris Speices")
plt.xlabel("Sepal Length(cm)")
plt.ylabel("Petal Length(cm)")
plt.legend()

#Displaying the plot
plt.show()
```

### **Task 07: Analyze Class Divisions**  

```python
#boxplots summarize 5 kinds of data: minimum, first quartile, median, third quartile, and maximum
#to utilize boxplots, need to gather the petal lengths by class in a list
#[setosa, versicolor, virginica] and call those in plt.boxplot()
#using [target == 'index number'] to filter the data

#grouping the petal lengths together by class
setosa = y_petal_length[target == 0]
versicolor = y_petal_length[target == 1]
virginica = y_petal_length[target == 2]

petal_class = [setosa, versicolor, virginica]

#now we can call this in plt.boxplot()
plt.boxplot(petal_class)

#labels and titles
plt.title("Box plot of the petal lengths by speices")
plt.xlabel("Species")
plt.ylabel("Petal Length(cm)")
plt.grid(True)

#using plt.xticks to replace the numbers with the species name
plt.xticks([1, 2, 3], ['setosa', 'versicolor', 'virginica'])

plt.show()
```

### **Task 8: Compute Correlations**

```python
#calculating the correlation coefficients (cf) between the features (using numpy)
#np.corrcoef() calculates the cf between rows by default, however the features are in columns

#using .T to transpose the features and turn them into rows
cf = np.corrcoef(features.T)

#setting a figure size to make it bigger
plt.figure(figsize=(8, 6))

#using matplotlibs .imshow to make a heatmap
plt.imshow(cf, cmap='coolwarm', interpolation='none')
plt.colorbar(label='Correlation Coefficient')

#annotating the heatmap
plt.title('Iris Features Correlation Heatmap')
#show all ticks and label them with the respective list entries
plt.xticks(ticks=np.arange(len(features_names)), labels=features_names, rotation=45)
plt.yticks(ticks=np.arange(len(features_names)), labels=features_names)

plt.show()
```

### **Task 9: Fit a Simple Linear Regression**

```python
#calculate the coefficient of determination R^2
#and then re-create the plot from task 6 and annotate it with the R^2 value

#importing r2_score for calculations
from sklearn.metrics import r2_score

#getting the R^2 value of the features used in task 5 and 6
r2 = r2_score(y_petal_length, line)

print(r2)
```

```python
#recreating the plot from task 6

#making the plot bigger
plt.figure(figsize=(8, 6))

#setting the colors for each class
colors = ['red', 'green', 'blue']

#plotting the points for each class using a for loop and range()

for target_class in range(3):
    plt.scatter(
        x_sepal_length[target == target_class],
        y_petal_length[target == target_class],
        color = colors[target_class],
        label = target_names[target_class]
    )

#adding a line of regression using .polyfit()
plt.plot(x_sepal_length, line, color = "skyblue", label = "Regression Line")

#annotating with R^2 value using .text()
plt.text(4.5, 6.5, f"$R^2$ = {r2:.3f}")

#labels, titles, and legend
plt.title("Sepal Length and Petal Length by Iris Speices")
plt.xlabel("Sepal Length(cm)")
plt.ylabel("Petal Length(cm)")
plt.legend()

#Displaying the plot
plt.show()
```

### **Task 10: Too Many Features**

```python
#using seaborn to make a pairplot of the dataset

import seaborn as sns

#for seaborn, the dataset has to be converted to a pandas dataframe
#can achieve this by using pd.DataFrame with the data and names of columns as the arguments
df = pd.DataFrame(data= iris_data.data, columns= features_names)

#using pd.Categorical to convert the class labels into readable text instead of an index
df['species'] = pd.Categorical.from_codes(target, target_names)

#making the pairplot, using hue to colorcode the data by the species
sns.pairplot(df, hue='species')

#displaying the plot
plt.show()
```

From what I can see, a pairplot pairs the features data together and creates a series of scatter plots and histograms for each of them, similar to what was made in task 05. It can be used to visualize the relationship between multiple variables.
