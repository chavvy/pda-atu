**Functions used throughout the notebook**  
Any created functions will be put here and called upon later throughout the notebook

```python
import numpy as np

#Functions for printing information on the iris dataset just for easier reading
def iris_keys(data):
    keys = data.keys()
    print(keys)

def iris_features(data):
    features = data['data']
    print(f"\nFeatures of the data:")
    print(features)

def iris_features_shape(data):
    features_shape = data['data'].shape
    print(f"\nShape of the data:")
    print(features_shape)

def iris_target(data):
    target = data['target']
    print(f"\nTarget of the data:")
    print(target)

def iris_shape(data):
    target_shape = data['target'].shape
    print(f"\nTarget shape of the data:")
    print(target_shape)

def iris_target_names(data):
    target_names = data['target_names']
    print(f"\nTarget names of the data:")
    print(target_names)

def iris_features_names(data):
    features_names = data['feature_names']
    print(f"\nFeature names of the data:")
    print(features_names)

#functions for getting the first and last rows of a dataset for task02
#using variable n in case there's a need to get first or last 'n' number of rows instead of 5
def iris_firstrow_data(data, n=5):
    features = data['data']
    print(f"\nFirst {n} rows of data:")
    print(features[:n])

def iris_lastrow_data(data, n=5):
    features = data['data']
    print(f"\nLast {n} rows of data:")
    print(features[-n:])

#functions for summarizing the iris dataset for task 03
def iris_feature_means(data):
    features = data['data']
    feature_names = data['feature_names']
    print("\nMean of each feature:")
    for r, name in enumerate(feature_names):
        print(f"{name}: {np.mean(features[:, r]):.2f}")

def iris_feature_mins(data):
    features = data['data']
    feature_names = data['feature_names']
    print("\nMinimum of each feature:")
    for r, name in enumerate(feature_names):
        print(f"{name}: {np.min(features[:, r]):.2f}")

def iris_feature_maxs(data):
    features = data['data']
    feature_names = data['feature_names']
    print("\nMaximum of each feature:")
    for r, name in enumerate(feature_names):
        print(f"{name}: {np.max(features[:, r]):.2f}")

def iris_feature_stds(data):
    features = data['data']
    feature_names = data['feature_names']
    print("\nStandard deviation of each feature:")
    for r, name in enumerate(feature_names):
        print(f"{name}: {np.std(features[:, r]):.2f}")

def iris_feature_medians(data):
    features = data['data']
    feature_names = data['feature_names']
    print("\nMedian of each feature:")
    for r, name in enumerate(feature_names):
        print(f"{name}: {np.median(features[:, r]):.2f}")
```

### **Task 01: Source the Data Set**

```python
from sklearn.datasets import load_iris

#loading iris dataset
iris_data = load_iris()

#The Iris dataset contains information on different samples of iris flowers, first used by Fisher
print(iris_data)
```

### **Task 02: Explore the Data Structure**

```python
#The data has 150 iris flowers in it, with each having 4 features, as shown by the shape
iris_features_shape(iris_data)
#This showcases the first 5 rows of data, and the last 5 rows of data
iris_firstrow_data(iris_data)
iris_lastrow_data(iris_data)
#The features are the sepal length and width, and the petal length and width (4 total)
iris_features_names(iris_data)
#The flowers are categorized into 3 different spieces: setosa, versicolor, virginica
iris_target_names(iris_data)
```

### **Task 03: Summarize the Data**

```python
#Summarizing the iris dataset (using numpy)

iris_feature_means(iris_data)
iris_feature_mins(iris_data)
iris_feature_maxs(iris_data)
iris_feature_maxs(iris_data)
iris_feature_stds(iris_data)
iris_feature_medians(iris_data)
```

### **Task 04: Visualize Features**

```python
#plotting a histogram for each feature in the data
import matplotlib.pyplot as plt

h_features = iris_data['data']
h_features_names = iris_data['feature_names']

#making a loop to go through each column
for r in range(h_features.shape[1]):
    #plotting the histogram
    plt.hist(h_features[:, r], bins=15, alpha=0.6, color= "royalblue", edgecolor = "black", label = "Iris Flowers")

    #labels, titles, and legend
    plt.title(f"Histogram of {h_features_names[r]}")
    plt.xlabel(h_features_names[r])
    plt.ylabel("Frequency")
    plt.legend()

    #displaying the plot
    plt.show()

```
